```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
hue_size = 8
brightness_size = 8
base_colors = sns.husl_palette(hue_size, l=0.85)

axes: list[list[plt.Axes]]
fig, axes = plt.subplots(hue_size, brightness_size, figsize=(hue_size, brightness_size), dpi=300)
plt.subplots_adjust(wspace=0, hspace=0.1)

cells = [[0 for _ in range(brightness_size)] for __ in range(hue_size)]

for i in range(hue_size):
    for j in range(brightness_size):
        color = np.array(base_colors[i]) - (j / (brightness_size + 3)) * np.array(base_colors[i])
        display_color = np.array(color * 255, dtype=int)
        axes[i, j].bar(0, 1, color=color)
        axes[i, j].axis('off')
        cells[i][j] = [float(color[0]), float(color[1]), float(color[2])]
        text_color = "black" if j < brightness_size - 3 else "white"
        axes[i, j].text(0, 0.5, f"R {display_color[0]}\nG {display_color[1]}\nB {display_color[2]}", ha='center', va='center', fontsize=8, color=text_color)
plt.show()
with open("./color_lighter.json", "w") as f: json.dump(cells, f)
```
